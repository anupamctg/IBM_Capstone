```python
column_names = ['Condo_ID','Condo_Name','Condo_lat','Condo_lon','School_name','School_lat','School_lon','Distance(Meter)']
df = pd.DataFrame(columns = column_names)

for i,a,b,c in zip(df_condos['id'], df_condos['name'], df_condos['location.lat'], df_condos['location.lng']):
    loc1=(b,c)
    for x,y,z in zip(df_schools['name'], df_schools['location.lat'], df_schools['location.lng']):
        loc2=(y,z)
        p=hs.haversine(loc1,loc2)
        if p<=4:
            df = df.append({'Condo_ID':i,
                        'Condo_Name':a,
                        'Condo_lat':b,
                        'Condo_lon':c,
                        'School_Name':x,
                        'School_lat':y,
                        'School_lon':z,
                        'Distance(Meter)':p*1000}, 
                       ignore_index=True)

df.dropna(axis=1,inplace=True)
df.sort_values(by=['Distance(Meter)'],inplace=True)
```
